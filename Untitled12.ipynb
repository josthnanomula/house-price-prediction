```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Assuming 'data' is your DataFrame
x_columns = ["POSTED_BY", "UNDER_CONSTRUCTION", "RERA", "BHK_NO.", "BHK_OR_RK", "SQUARE_FT", "READY_TO_MOVE", "RESALE", "LONGITUDE", "LATITUDE"]
x = data[x_columns].values
y = data["TARGET(PRICE_IN_LACS)"].values

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Machine": SVR(),
    "K-Nearest Neighbors": KNeighborsRegressor()
}

# Train and evaluate each model
mse_results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred,squared=False)
    mse_results[name] = mse

# Print MSE for each model
for name, mse in mse_results.items():
    print(f"{name}: {mse}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, VotingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import StackingRegressor
import numpy as np

# Assuming 'data' is your DataFrame
x_columns = ["POSTED_BY", "UNDER_CONSTRUCTION", "RERA", "BHK_NO.", "BHK_OR_RK", "SQUARE_FT", "READY_TO_MOVE", "RESALE", "LONGITUDE", "LATITUDE"]
x = data[x_columns].values
y = data["TARGET(PRICE_IN_LACS)"].values

# Splitting data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define regression models
linear_reg = LinearRegression()
ridge_reg = Ridge()
lasso_reg = Lasso()
decision_tree_reg = DecisionTreeRegressor()
random_forest_reg = RandomForestRegressor()
gradient_boosting_reg = GradientBoostingRegressor()
knn_reg = KNeighborsRegressor()
adaboost_reg = AdaBoostRegressor()
xgb_reg = XGBRegressor()

# Ensemble Voting
voting_reg = VotingRegressor(estimators=[
    ('linear', linear_reg),
    ('ridge', ridge_reg),
    ('lasso', lasso_reg),
    ('decision_tree', decision_tree_reg),
    ('random_forest', random_forest_reg),
    ('gradient_boosting', gradient_boosting_reg),
    ('knn', knn_reg),
    ('adaboost', adaboost_reg),
    ('xgb', xgb_reg)
])

# Stacking
stacking_reg = StackingRegressor(estimators=[
    ('linear', linear_reg),
    ('ridge', ridge_reg),
    ('lasso', lasso_reg),
    ('decision_tree', decision_tree_reg),
    ('random_forest', random_forest_reg),
    ('gradient_boosting', gradient_boosting_reg),
    ('knn', knn_reg),
    ('adaboost', adaboost_reg),
    ('xgb', xgb_reg)
], final_estimator=RandomForestRegressor())

# Train and evaluate models
models = {
    'Linear Regression': linear_reg,
    'Ridge Regression': ridge_reg,
    'Lasso Regression': lasso_reg,
    'Decision Tree Regression': decision_tree_reg,
    'Random Forest Regression': random_forest_reg,
    'Gradient Boosting Regression': gradient_boosting_reg,
    'KNN Regression': knn_reg,
    'AdaBoost Regression': adaboost_reg,
    'XGBoost Regression': xgb_reg,
    'Ensemble Voting': voting_reg,
    'Stacking': stacking_reg
}

results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    n = len(y_test)
    p = len(x_columns)
    adj_r2 = 1 - (1 - r2) * ((n - 1) / (n - p - 1))
    results[name] = {'R^2': r2, 'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'Adjusted R^2': adj_r2}

# Print results for each model
for name, metrics in results.items():
    print(f"Metrics for {name}:")
    print(f"R^2: {metrics['R^2']:.4f}")
    print(f"MSE: {metrics['MSE']:.4f}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"MAE: {metrics['MAE']:.4f}")
    print(f"Adjusted R^2: {metrics['Adjusted R^2']:.4f}")
    print("\n")
```
